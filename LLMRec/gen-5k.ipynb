```python
def does_user_item_pair_exists(user_id: int, item_id: int) -> bool:
    user_item_pair_row_df = data_df.loc[(data_df['user_id'] == user_id) & (data_df['item_id'] == item_id)]
    return False if user_item_pair_row_df.empty else True

```

```python
import random
```

```python
def get_n_new_user_item_pairs(user_id: int, n: int) -> list[int]:
    new_item_ids: list[int] = []
    for i in range(n):
        rdm_item_id = random.randrange(1, item_count+1)
        while (does_user_item_pair_exists(user_id, rdm_item_id)):
            rdm_item_id = random.randrange(1, item_count+1)
        new_item_ids.append(rdm_item_id)
    return new_item_ids

first_user_pairs_random = get_n_new_user_item_pairs(1, 5)
print(first_user_pairs_random)
```

```python
def get_item_information_by_id(item_id: int) -> dict[str, any]:
    item_info: dict[str, any] = {}
    item_row_df = item_df.loc[item_df['movie_id'] == item_id]

    for key in ['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 
    'unknown', 'Action', 'Adventure', 'Animation', "Children's", 'Comedy', 'Crime',
    'Documentary', 'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance',
    'Sci-Fi', 'Thriller',  'War', 'Western']:
        item_info[key] = item_row_df[key].item()

    return item_info

item_information = get_item_information_by_id(2)
print(item_information)
```

```python
def get_user_information_by_id(user_id: int) -> dict[str, any]:
    user_info: dict[str, any] = {}
    user_row_df = user_df.loc[user_df['user_id'] == user_id]

    for key in ['user_id', 'age', 'gender', 'occupation', 'zip_code']:
        user_info[key] = user_row_df[key].item()

    return user_info

user_information = get_user_information_by_id(2)
print(user_information)
```

```python
def get_movies_by_user_id(user_id: int) -> dict[int, int]:
    movie_id_2_rating: dict[int, int] = {}
    user_movies_df = data_df.loc[data_df["user_id"] == user_id]
    for index, row in user_movies_df.iterrows():
        movie_id_2_rating[row["item_id"]] = row["rating"] 
    return movie_id_2_rating

movies_for_user_one = get_movies_by_user_id(1)
print(movies_for_user_one)
```

```python
# Generate a bunch of information about user's past history and profile to predict a new movie rating

def predict_user_movie_rating(user_info: dict[str, any], user_ctx: dict[str, any], new_movie_info: dict[str, any]) -> str:
    # user_ctx: dictionary containing movies seen, their info, and rating
    # movie_info: {'movie_id': 2, 'movie_title': 'GoldenEye (1995)', 'release_date': '01-Jan-1995', 'video_release_date': nan, 'IMDb_URL': 'http://us.imdb.com/M/title-exact?GoldenEye%20(1995)', 'unknown': 0, 'Action': 1, 'Adventure': 1, 'Animation': 0, "Children's": 0, 'Comedy': 0, 'Crime': 0, 'Documentary': 0, 'Drama': 0, 'Fantasy': 0, 'Film-Noir': 0, 'Horror': 0, 'Musical': 0, 'Mystery': 0, 'Romance': 0, 'Sci-Fi': 0, 'Thriller': 1, 'War': 0, 'Western': 0}
    # user_info: {'user_id': 2, 'age': 53, 'gender': 'F', 'occupation': 'other', 'zip_code': '94043'}

    def concat_user_info(user_info: dict[str, any]) -> str:
        user_string = "User with "
        user_string += f"Age f{user_info['age']}, "
        user_string += f"Gender f{user_info['gender']}, "
        user_string += f"Occupation f{user_info['occupation']}, "
        user_string += f"and Zip Code f{user_info['zip_code']}"
        return user_string

    def concat_movie_info(movie_info: dict[str, any], rating: int):
        movie_string = "watched and liked movie titled "
        movie_string += f"'{movie_info['movie_title']}' "
        movie_string += f"released on {movie_info['release_date']} "
        if (movie_info["unknown"]):
            movie_string += "has no genres "
        else:
            movie_string += "is under the genres "
            for genre in ["Action", "Adventure", "Animation", "Children's", "Comedy", "Crime", "Documentary", "Drama", "Fantasy"
                          , "Film-Noir", "Horror", "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", "Western"]:
                if movie_info[genre]:
                    movie_string += f"{genre} "
        if (rating != -1): # rating exists
            movie_string += f"and rated it {rating}"
        return movie_string

    user_string: str = f"{concat_user_info(user_info)}\n"
    movies_string: str = "has watched \n"
    for user_movie_id in user_ctx.keys():
        movie_info = user_ctx[user_movie_id]["movie_info"]
        rating = user_ctx[user_movie_id]["rating"]
        movies_string += f"{concat_movie_info(movie_info, rating)}\n"
    
    new_movie_string = f"Predict what the user will rate (out of 5) the following movie:\n{concat_movie_info(new_movie_info, -1)}\n"

    instruction_string = "Predict what the user will rate the movie on a scale from 1 (low) to 5 (high). Explain your answer based on the user's movie history. At the end, write a single number from 1 to 5 but put it at the very of your answer. Put the rating at the end in the format ::5:: but with your rating in the place of 5. Do not be optimistic. Be a harsh critic and consider a rating of 1 to be the most common rating."

    combined_string = user_string + movies_string + new_movie_string + instruction_string

    response = ollama.chat(model='llama2', messages=[
    {
        'role': 'user',
        'content': combined_string,
    },
    ])
    
    return response['message']['content']
    
```

```python
user_item_ratings: list[tuple[str, str, str]] = []

for i in range(1, user_count+1):
    user_info = get_user_information_by_id(i)
    user_movies = get_movies_by_user_id(i)
    user_ctx = {}
    for user_movie_id in user_movies.keys():
        movie_info = get_item_information_by_id(user_movie_id)
        user_ctx[user_movie_id] = {"movie_info": movie_info, "rating": user_movies[user_movie_id]}
    five_new_movies = get_n_new_user_item_pairs(i, 5)
    for movie_id in five_new_movies:
        new_movie_info = get_item_information_by_id(movie_id)
        rating_answer: str = predict_user_movie_rating(user_info, user_ctx, new_movie_info)
        rating_regex: str = re.search("::(\d)::", rating_answer)
        if rating_regex is not None:
            user_id, movie_id, rating = user_info["user_id"], movie_id, rating_regex.group(1)
            #print(user_id, movie_id, rating)
            print(user_id)
            user_item_ratings.append([user_id, movie_id, rating])
        
```

```python
print(user_item_ratings)
filename = "user_item_predictions.txt"


with open(filename, "w") as f:
    f.write("")

with open(filename, "a") as f:
    for user_item_pair in user_item_ratings:
        f.write(f"{user_item_pair[0]}, {user_item_pair[1]}, {user_item_pair[2]}\n")
```
